####Dynamic notebook

```python
dbutils.widgets.text("file_name","")
```

```python
p_file_name=dbutils.widgets.get("file_name")
```

```python
p_file_name
```

####Data Reading

```python
df=spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format","parquet")\
        .option("cloudFiles.schemaLocation",f"abfss://bronze@datalakedatabricksram.dfs.core.windows.net/checkpoint_{p_file_name}")\
            .load(f"abfss://source@datalakedatabricksram.dfs.core.windows.net/{p_file_name}")
```

####Data Writing

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
        .option("checkpointLocation",f"abfss://bronze@datalakedatabricksram.dfs.core.windows.net/checkpoint_{p_file_name}")\
            .option("path",f"abfss://bronze@datalakedatabricksram.dfs.core.windows.net/{p_file_name}")\
                .trigger(once=True)\
                    .start()
```

```python
df=spark.read.format("parquet")\
    .load(f"abfss://bronze@datalakedatabricksram.dfs.core.windows.net/{p_file_name}")
display(df)
```
